# Objective

Our Objective is to find the "Employee Salary Negotiation with respect to Years of Experience". Whether he is saying right salary for the specified experience. 

It can be used to identify the bluff which a new employee can made on his past job salary. The model intakes an experience level and predict the right salary for a given experience / position level.

# Polynomial Regression

In Polynomial Regression, First We transform our features into Polynomial form (X, X^2, X^3, . . .) then we inject the polynomial formed data to the Linear Model.

Fit -> It Calculates the equation needed elements (X^2, X^3 . . ) 
Transform --> Then It transforms the equation by fitting the elements.

Polynomial Regression is also a type of Linear Regressio. Because, linear represents the Coefficients not the features. See only the Polynomial features are having huge powers not coefficients (m & c)



y= m1x + m2x^2 + m3x^3 + m4x^4 + m5x^5 + c

### Importing the Packages

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

import seaborn as sns
import matplotlib.pyplot as plt
```

### Reading the Dataset

```python
df = pd.read_csv("Salary_Data.csv")

df.head()
```

### Plotting the Values

```python
X = df['YearsExperience'].values
y = df['Salary'].values

X_loc = df.iloc[:,0:1].values
y_loc = df.iloc[:,1:2].values

plt.scatter(X_loc, y_loc)
```

First fitting the Best fit line using Linear Regression and then use Polynomial Regression

### Visualizing the distributions

```python
sns.distplot(X_loc)
```

```python
sns.distplot(y_loc)
```

### Linear Model

```python
linearmodel = LinearRegression()

linearmodel.fit(X_loc, y_loc)
```

```python
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Experience vs Salary Linear Relationship")
plt.scatter(X_loc, y_loc)


plt.plot(X_loc, linearmodel.predict(X_loc), color = 'red')
```

### Polynomial Regression

In this Polynomial Regression, We try to transform the X Features to the Polynomial form (i.e X^2, X^3) and then fit into the Linear Regression. Usually by default the degree of polynomial is 2.

```python
X_polynomial = PolynomialFeatures(degree=6)
x_poly = X_polynomial.fit_transform(X_loc)
```

```python
linearmodel_poly = LinearRegression()

linearmodel_poly.fit(x_poly, y_loc)
```

### Plotting the Polynomial Features Converted Data

```python
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Experience vs Salary Linear Relationship using Polynomial Features")
plt.scatter(X_loc, y_loc)
plt.plot(X_loc, linearmodel_poly.predict(x_poly), color = 'red')
```

### Regression Accuracy Calculation using Mean Squared Error(MAE)

Lesser the MAE, the more the Accuracy of the Predicted Results

```python
# Mean Squared Error
from sklearn.metrics import mean_squared_error

mean_squared_error(y_loc, linearmodel_poly.predict(x_poly))
```
